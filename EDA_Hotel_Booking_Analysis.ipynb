## 3. ***Data Wrangling***

### Data Wrangling Code

**DATA CLEANING:** The first and the most important step of Data Analysis is Data cleaning which means making the data ready for analysis.
In Data Cleaning we remove the unwanted elements from our Data set so that they do not cause disturbance in the accuracy of our results.

**STEPS FOLLOWED:**

**STEP 1:** *Creating a copy*

**STEP 2:** *Deleting the duplicate values*

**STEP 3:** *Dealing with the null values*

**STEP 4:** *Converting columns to correct datatype*

**STEP 5:**  *Adding new columns in the data*

**STEP 6:** *Creating subsets from the original dataset to be used in Data Analysis*


**STEP 1:** ***CREATING A COPY***

````python
# Write your code to make your dataset analysis ready.
#Creating a copy of the dataset to keep the original data intact
hotel_booking=hotel_data.copy()
````

**STEP 2:** ***DELETING THE DUPLICATE VALUES***

````python
#Delete all the duplicate values
hotel_booking.drop_duplicates(inplace=True)
hotel_booking.shape #count of rows after deleting the duplicate values
````

**STEP 3:** ***DEALING WITH THE NULL VALUES***

````python
#Dealing with the Null values
hotel_booking.isnull().sum().sort_values (ascending=False) [:6] #checking all the columns with null values
````

````python
hotel_booking['company'].fillna(0, inplace=True) #Filling the Null values in the column company by 0
hotel_booking['agent'].fillna(0, inplace=True) #Filling the Null values in the column agent by 0
hotel_booking['children'].fillna(hotel_booking['children'].mean(), inplace = True) #Column children also has 0 as a value so replacing the null value with Mean 
hotel_booking['country'].fillna('others', inplace=True) #Column country is of type string so we have replaced it with the string value
````

````python
hotel_booking.isnull().sum().sort_values (ascending=False) [:6] #checking if there are any null values remaining
````

**STEP 4:** ***CONVERTING COLUMNS TO CORRECT DATATYPES***

````python
hotel_booking.info()
````

````python
#converting the datatype from float64 to int64 for columns 'company','agent','children'.
hotel_booking[['company','agent','children']]= hotel_booking[['company','agent','children']].astype('int64') 
#converting datatype of column 'reservation_status_date' to date_type.
hotel_booking['reservation_status_date'] = pd.to_datetime(hotel_booking['reservation_status_date'], format ='%Y-%m-%d') 
````

**STEP 5:** ***ADDING NEW COLUMNS IN THE DATA AND DROPPING ROWS***


````python
hotel_booking['total_people']= hotel_booking['adults']+hotel_booking['children']+hotel_booking['babies'] #a column that gives total number of people that stayed in the hotel
hotel_booking['total_stay']= hotel_booking['stays_in_weekend_nights']+hotel_booking['stays_in_week_nights'] # a column that gives total days of stay in the hotel
hotel_booking[hotel_booking['adults']+hotel_booking['children']+hotel_booking['babies']==0] #Deleting all the rows where the total number of people=0
hotel_booking.drop(hotel_booking[hotel_booking['adults']+hotel_booking['children']+hotel_booking['babies']==0].index, inplace=True) #166 rows dropped
````

**STEP 6**: ***CREATING SUBSETS FROM THE ORIGINAL DATSET TO USE IN EDA***

````python
not_canceled_data= hotel_booking[hotel_booking['is_canceled']==0] #a dataset where the bookings were not canceled
canceled_data= hotel_booking[hotel_booking['is_canceled']==1] #a dataset where the bookings were canceled
````

#### Chart - 1 -   Booking Percentage (Univariate)

**Q.1 WHAT IS THE PERCENTAGE OF BOOKING FOR EACH HOTEL??**

````python
hotel_booking['hotel'].count() #total bookings made
hotel_booking['hotel'].value_counts() #bookings made for particular hotel 

#calculating the percentage by dividing the particular hotel count by total bookings
hotel_percentage=pd.DataFrame(hotel_booking['hotel'].value_counts()/hotel_booking['hotel'].count()*100)\
.reset_index().rename(columns={'index':'Hotel Type', 'hotel':'Booking Percentage'}) 

# Pie Chart
data_Q1=hotel_percentage['Booking Percentage'].tolist() #converting the column Booking Percentage to list
labels_Q1=hotel_percentage['Hotel Type'].tolist() #converting the column Hotel Type to list
colors_Q1=['lightskyblue','yellow'] #setting the colors
explode=(0,0.1)
hotel_percentage
````

````python
# Chart - 1 visualization code
plt.rcParams['figure.figsize'] = (8,8)
plt.pie(data_Q1, labels=labels_Q1, autopct='%1.1f%%',shadow=True, colors=colors_Q1, explode=explode, startangle=90, textprops={'fontsize': 15})
plt.title("BOOKING PERCENTAGE FOR CITY AND RESORT HOTELS",fontsize=15,fontweight='bold')
plt.legend()
plt.show()
````

##### 1. Why did you pick the specific chart?


1) **A pie chart** expresses a part-to-whole relationship in the data.

Pie charts are used to represent the proportional data or relative data in a single chart. The concept of pie slices is used to show the percentage of a particular data from the whole pie

Pie chart is used where the values adds up to 100%.

2) In this chart City Hotel and Resort Hotel were needed to be compared by their booking percentage showing a part-to-whole relationship thus Piechart was used.

##### 2. What is/are the insight(s) found from the chart?



1) 61% Bookings for City Hotel and 39% of bookings for Resort Hotel were made.

2) More bookings were made for City Hotel as compared to Resort Hotel.
